## 1. Reshape (Changer la forme)
Très utile en Machine Learning pour réorganiser des données. Par exemple, transformer un tableau de 12 éléments en une matrice 3x4.

```python
import numpy as np
a = np.arange(25) # Crée un tableau de 0 à 11
print("Tableau 1D initial :", a)

b = a.reshape(-1, 5) # 3 lignes, 4 colonnes
print("\nMatrice 3x4 :\n", b)
```

## 2. Masques booléens (Filtrage)
Le masque booléen consiste à filtrer un tableau en utilisant une condition. Cela renvoie un tableau de `True` et `False`, qu'on peut appliquer au tableau pour ne garder que les `True`.

```python
ages = np.array([15, 22, 18, 30, 16, 25])

# Qui est majeur ?
majeur = ages >= 18
print("Masque booléen :", majeur)

# Appliquons le masque
print("Ages majeurs :", ages[majeur])
```

1. Créez un tableau `notes` contenant : [8, 12, 15, 9, 17, 5, 11].
2. Utilisez un masque booléen pour créer un nouveau tableau `notes_valides` ne contenant que les notes supérieures ou égales à 10.
3. Affichez `notes_valides`.

```python
# Écrivez votre code pour le défi ici
notes = np.array([8, 12, 15, 9, 17, 5, 11, 8])
notes_sup_10 = notes >= 10
print(f"Notes : {notes}")
print(f"Masque de filtrage : {notes_sup_10}")
print(f"Notes valides : {notes[notes_sup_10]}")
print(f"Matrice des notes valides :\n{notes[notes_sup_10].reshape(-1,2)}")
```
