# Jour 9 - 24 Juillet 2026 : Manipulation de Données avec Pandas

**Pandas** est l'outil incontournable pour analyser des données tabulaires (comme des fichiers Excel ou CSV). Il utilise deux structures principales : les **Series** (1 colonne) et les **DataFrames** (un tableau avec des lignes et des colonnes).

```python
import pandas as pd

# Création d'un DataFrame (tableau) à partir d'un dictionnaire
donnees = {
    'Nom': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Ville': ['Paris', 'Lyon', 'Marseille']
}

df = pd.DataFrame(donnees)
df
```

## Sélection de données
On peut sélectionner une seule colonne avec `df['NomColonne']` ou plusieurs avec une liste de colonnes.

```python
# Sélectionner seulement la colonne Age
print(df['Age'])

# Filtrer les personnes de plus de 28 ans
plus_de_28 = df[df['Age'] > 28]
plus_de_28
```
