## 3️⃣ Valeurs manquantes

Pandas encode les valeurs manquantes avec `NaN`. **On ne présume jamais** qu'un
dataset en contient ou non — on **vérifie systématiquement**, même quand tout
semble propre au premier coup d'œil.

````python
df.isna().sum().sort_values(ascending=False)
````

````python
# En pourcentage, c'est souvent plus parlant (surtout sur de gros datasets)
(df.isna().mean() * 100).round(1).sort_values(ascending=False)
````

❓ **Discussion avec le groupe** : le résultat est 0 partout. Est-ce que ça veut
dire qu'on peut sauter cette étape la prochaine fois ? **Non** — c'est justement
en la faisant systématiquement qu'on peut affirmer avec certitude qu'il n'y a rien
à traiter ici. Un Data Analyst qui saute la vérification sous prétexte que "ça a
l'air propre" prend un risque.

**Pour mémoire, si on avait trouvé des `NaN`, les 3 stratégies classiques auraient été :**

1. **Supprimer** la ligne (`dropna`) — si peu de lignes concernées et donnée critique
2. **Remplacer** par une valeur par défaut (`fillna`) — moyenne, médiane, mode, ou 0
3. **Laisser tel quel** et le signaler — si certains modèles savent gérer les `NaN`

*(pas nécessaire ici, mais gardez ces 3 options en tête pour vos futurs projets)*

## 4️⃣ Nettoyage — audit qualité

Même sans `NaN`, un Data Analyst sérieux contrôle toujours :
- les **types de colonnes** (un nombre stocké en texte, par exemple)
- le **texte mal formaté** (espaces, casse incohérente)
- les **doublons**
- les **valeurs aberrantes** (outliers)

On va passer ce dataset au crible, point par point.

````python
# Les types sont-ils cohérents avec ce qu'on attend ?
# (age, failures, notes... doivent être numériques ; sex, school... du texte)
df.dtypes
````

````python
# Zoom sur deux colonnes sensibles : age et sex
print(df["age"].unique())
print(df["sex"].unique())
````

✅ **Résultat de l'audit types/texte** : `age` est bien numérique (pas de "18 ans"
en texte), `sex` ne contient que des valeurs propres (`F`/`M`, pas d'espace ni de
casse incohérente). Rien à corriger — mais on a vérifié, on ne l'a pas supposé.

````python
# Doublons : des lignes strictement identiques ?
df.duplicated().sum()
````

✅ **0 doublon.** Là encore, le réflexe (`.duplicated().sum()`) est plus important
que le résultat : sur un dataset de plusieurs milliers de lignes, des doublons
passent facilement inaperçus si on ne les cherche pas explicitement.

````python
# Valeurs aberrantes potentielles : on regarde la distribution de 'absences'
df["absences"].describe()
````

````python
# On regarde les valeurs extrêmes en détail
df.sort_values("absences", ascending=False).head(5)[["age", "sex", "absences"]]
````

❓ **Discussion avec le groupe** : une valeur de 75 absences interpelle — est-ce
une erreur de saisie ou une situation réelle (décrochage, maladie longue, etc.) ?
**Sans autre information, on ne modifie pas une donnée réelle sur une simple
intuition.** Face à un vrai outlier suspect, le bon réflexe est de se renseigner
(documentation du dataset, personne qui a collecté les données) avant de corriger
quoi que ce soit — supprimer ou plafonner une valeur légitime fausserait l'analyse.
Ici, on **garde la valeur telle quelle** et on la garde en tête pour la suite
(elle pourra par exemple justifier une standardisation avant le Machine Learning).

## 5️⃣ Filtrage

On sélectionne des sous-ensembles de lignes selon des conditions — le pain quotidien
d'un Data Analyst.


````python
# Les étudiants qui veulent poursuivre des études supérieures ET ont internet à la maison
filtre = (df["higher"] == "yes") & (df["internet"] == "yes")
df[filtre].shape[0]
````

````python
# Les étudiants en difficulté : au moins un échec précédent ET note finale < 10
en_difficulte = df[(df["failures"] > 0) & (df["G3"] < 10)]
en_difficulte[["age", "sex", "studytime", "failures", "G3"]].head()
````

## 6️⃣ Tri

`sort_values` pour classer les étudiants selon un critère.


````python
# Top 5 des meilleures notes finales
df.sort_values("G3", ascending=False)[["age", "sex", "studytime", "G3"]].head()
````

````python
# Tri multi-critères : par sexe, puis par note décroissante
df.sort_values(["sex", "G3"], ascending=[True, False]).head(8)[["sex", "G3"]]
````

## 7️⃣ Création de colonnes

On enrichit le dataset avec des colonnes dérivées — souvent la clé pour un bon modèle
de Machine Learning plus tard.


````python
# Moyenne des 3 notes (G1, G2, G3)
df["moyenne_generale"] = df[["G1", "G2", "G3"]].mean(axis=1).round(2)

# Une colonne catégorielle : l'étudiant a-t-il réussi ? (moyenne >= 10)
df["reussite"] = np.where(df["moyenne_generale"] >= 10, "admis", "non_admis")

# Temps d'écran social/loisir global (sorties + alcool weekend)
df["indice_social"] = df["goout"] + df["Walc"]

df[["G1", "G2", "G3", "moyenne_generale", "reussite", "indice_social"]].head()
````

## 8️⃣ Groupby & statistiques

`groupby` permet d'agréger les données par catégorie — indispensable pour un rapport
d'analyse.


````python
# Moyenne générale par sexe
df.groupby("sex")["moyenne_generale"].mean().round(2)
````

````python
# Taux de réussite par temps d'étude hebdomadaire (studytime)
df.groupby("studytime")["reussite"].value_counts(normalize=True).round(2)
````

````python
# Plusieurs statistiques d'un coup avec agg()
df.groupby("sex").agg(
    moyenne=("moyenne_generale", "mean"),
    absences_moy=("absences", "mean"),
    effectif=("moyenne_generale", "count"),
).round(2)
````
